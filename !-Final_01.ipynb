```python
#Вспомогательная функция для отображения меток данных на диаграмме
def show_values_on_bars(axs, h_v="v", space=0.4):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, value,fontsize=10, ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                try:
                    _x = p.get_x() + p.get_width() + float(space)
                    _y = p.get_y() + p.get_height()
                    value = int(p.get_width())
                    ax.text(_x, _y, value,fontsize=10, ha="left")
                except ValueError:
                    x =0
                    #print('Nan не выводим.')
    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
```

```python
#campaigns.csv - список кампаний с параметрами поиска по ним. 
#Если нет значения, например, в Пол, то подходит креатор с любым значением в этом поле.

campaigns = pd.read_csv('campaigns.csv')
#разделим campaign_categories на отдельные серии
campaigns['campaign_categories'] = campaigns['campaign_categories'].str.split(';')
campaigns.info()
```

```python
#подсчет категорий представленных в каждой кампании 
campaigns['campaign_categories_cnt'] = campaigns['campaign_categories'].apply(lambda x: x if x!=x else len(x))
#campaigns.head()
print('Максимальное количество категорий в одной кампании: ', campaigns['campaign_categories_cnt'].max())
```

```python
#преобразуем полученный файл в отдельный датафрейм
campaigns_update = pd.read_csv('campaigns_update.csv')
#campaigns_update.head()
campaigns_update.info()
```

```python
#выделим отдельную категорию "followers"
campaigns_update['gender'] = campaigns_update['gender'].fillna('None')
campaigns_update['followers_from'] = campaigns_update['followers_from'].fillna('0')
campaigns_update['followers_to'] = campaigns_update['followers_to'].fillna('и выше')
campaigns_update['followers'] = (campaigns_update['followers_from'].astype(str)+' - '
                                +campaigns_update['followers_to'].astype(str))
campaigns_update['followers_cat'] = campaigns_update['followers'].astype("category")
#campaigns_update['followers'].value_counts()
```

```python
#создадим отдельный список с перечислением всех кампаниий и их категорий
demands_list = []

for rowIndex, row in campaigns_update.iterrows(): #
    for r in row.index:
        if 'categories' in r:
            demands_list.append([row['campaign_id'], row['gender'], row['followers_cat'], row[r]])
print('Всего записей: ', len(demands_list))
print('474 кампании * 19 возможных категорий = ', 474*19)
```

```python
#преобразуем полученный список в датафрейм
demands_df = pd.DataFrame(demands_list)
demands_df = demands_df.rename(columns={0:'campaign_id', 1:'gender',2:'followers_cat', 3:'category_name'})
#demands_df.head()
demands_df.info()
```

```python
#посчитамем сколько кампаний имеется в каждой из категорий
demands_group_df = demands_df.groupby(['category_name','gender','followers_cat']).count().reset_index(drop=False)#.sort_values('start_time')
#demands_group_df = demands_df.groupby(['category_name']).count().reset_index(drop=False)#.sort_values('start_time')


demands_group_df = demands_group_df.rename(columns={'campaign_id':'campaign_count'})
#demands_group_df.head()
demands_group_df.info()
```

```python
#Сформируем сводную таблицу, в которой посчитаем количество заказов от брендов по категориям
demands_group_pivot_df = demands_group_df.pivot_table(values=['campaign_count'],
                                                       index=['gender','followers_cat'],
                                                       columns=['category_name'],
                                                       aggfunc='sum')
demands_group_pivot_df.columns = list(map(lambda i: i[-1], demands_group_pivot_df))
demands_group_pivot_df = demands_group_pivot_df.reset_index(drop=False)
demands_group_pivot_df = demands_group_pivot_df.fillna(0)
#demands_group_pivot_df.to_excel('Спрос_от_брендов.xlsx')
demands_group_pivot_df.head()
```

```python
#creators_gender_followers_campaigns - список креаторов с параметрами Пола, Кол-ва подписчиков и списком кампаний, 
#где креатор проявил интерес. Можем считать, что “проявил интерес” - это значит спрос совпал с предложением

data = pd.read_csv('creators_gender_followers_campaigns_copy.csv', delimiter=';')

genders = pd.read_csv('genders.csv')
data.info()
```

```python
#вспомагательная функция для приведения gender_id к его названию
def gender_fill(g):
    if g.loc['gender_id'] == genders.loc[0, 'id']:
        return genders.loc[0, 'name']
    elif g.loc['gender_id'] == genders.loc[1, 'id']:
        return genders.loc[1, 'name']
    elif g.loc['gender_id'] == genders.loc[2, 'id']:
        return genders.loc[2, 'name']
    elif g.loc['gender_id'] == genders.loc[3, 'id']:
        return genders.loc[3, 'name']
    
data['gender'] = data.apply(gender_fill, axis=1)
data.head()
```

```python
#создадим отдельный список с перечислением всех креаторов и их заинтересовавших их кампаний
temp_list = []

for rowIndex, row in data.iterrows(): #
    for r in row.index:
        if 'campaigns' in r:
            temp_list.append([row['creator_id'],row['gender'], row['followed_by_count'], row[r]])

print('Всего записей: ', len(temp_list))
print('3140 креаторов * 158 кампаний = ', 3140*158)
```

```python
#преобразуем полученный список в датафрейм
temp_df = pd.DataFrame(temp_list)
temp_df = temp_df.rename(columns={0:'creator_id', 1:'gender', 2:'followed_by_count', 3:'campaign_id'})
temp_df = temp_df.dropna()
#temp_df.shape
temp_df.head()
```
